```python
import numpy as np
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])
#qn1
A_inv = np.linalg.inv(A)
print("A * A^-1 =\n", A @ A_inv)
#qn2
print("AB =\n", A @ B)
print("BA =\n", B @ A)
print("AB == BA?", np.array_equal(A @ B, B @ A))
#qn3
print("(AB)^T =\n", (A @ B).T)
print("B^T * A^T =\n", B.T @ A.T)
print("(AB)^T == B^T * A^T?", np.allclose((A @ B).T, B.T @ A.T))
```
