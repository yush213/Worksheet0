**3.3 Exploratory Data Analysis with Group-by Method Practice Exercise:**

```python
total_fare = titanic_new["Fare"].sum()
print("Total Fare collected:", total_fare)
```

```python
first_class_passengers = titanic_new[titanic_new["Pclass"] == 1]
total_fare_first_class = first_class_passengers["Fare"].sum()
print("Total Fare from first class passengers:", total_fare_first_class)
```

```python
second_class_passengers = titanic_new[titanic_new["Pclass"] == 2]
total_fare_second_class = second_class_passengers["Fare"].sum()
print("Total Fare from second class passengers:", total_fare_second_class)
```

```python
third_class_passengers = titanic_new[titanic_new["Pclass"] == 3]
total_fare_third_class = third_class_passengers["Fare"].sum()
print("Total Fare from third class passengers:", total_fare_third_class)
```

```python
fare_totals = [total_fare_first_class, total_fare_second_class, total_fare_third_class]
fare_totals
```

```python
fare_proportions = [fare / total_fare for fare in fare_totals]
fare_proportions
```

```python
def categorize_age(age):
    if pd.isna(age):
        return "unknown"
    elif age < 18:
        return "child"
    elif 18 <= age < 65:
        return "adult"
    else:
        return "senior"

titanic_new["age_group"] = titanic_new["Age"].apply(categorize_age)

print(titanic_new[["Name", "Age", "age_group"]].head())
```

```python
total_passengers = titanic_new.shape[0]
print("Total number of passengers:", total_passengers)
```

```python
age_group_counts = titanic_new["age_group"].value_counts()
age_group_counts
```

```python
age_group_proportions = age_group_counts / total_passengers
age_group_proportions
```

```python
age_group_percentages = age_group_proportions * 100
age_group_percentages
```

```python
age_group_percentages = (age_group_counts / total_passengers) * 100
print(age_group_percentages)
```
